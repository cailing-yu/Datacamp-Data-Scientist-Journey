````python
import pandas as pd

dogs = pd.read_csv('dogs.csv')

# Call the pivot_table method on the dogs DataFrame
# 1st argument: the column to aggregate (height)
# index: lists the columns to group by and display as rows(breed)
# columns: lists the columns to group by and display as columns (color)
# Default aggregation function is mean, so the resulting table will show the average height for each breed and color combination
dogs_height_by_breed_and_color = dogs.pivot_table('height', index='breed', columns='color')

````

```.pivot_table()```: DataFrames with sorted indexes.

```.loc[]``` + slicing combination is ideal for subsetting ```.pivot_table()```



````python
dogs_height_by_breed_and_color.loc['Chow Chow': 'Poodle']
````

```.mean(axis="")```

```.mean(axis='index')``` Default value is index: calculate the statistic across rows

````python
# Calculate the mean height for each breed across all colors
dogs_height_by_breed_and_color.mean(axis='index')
````

```.mean(axis='columns')``` : calculate the statistic across columns, for each row

````python
# Calculate the mean height for each breed across all colors
dogs_height_by_breed_and_color.mean(axis='columns')
````
