## 1. Closure with External State (外部状态闭包)
* **Key**: Side-effect driven search. Relies on an external container to constantly overwrite and store the latest valid ancestor.
* **Thought Process**: Start at root $\rightarrow$ move downwards based on BST properties $\rightarrow$ aggressively overwrite the external variable with the current node $\rightarrow$ stop when a divergence (split) occurs $\rightarrow$ extract the final value from the external container.
* **Pitfalls**: **Side Effects & Redundancy**. Mutating variables outside the local scope breaks pure function principles. Writing to memory at every intermediate node wastes compute resources.
* **Pattern**: State-driven DFS (An anti-pattern in modern engineering, to be avoided unless handling complex global back-tracking).

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        # Globale variable
        lca = [root]

        # Helper() to search for the lca
        def search(root):

            # Base Case
            if not root:
                return

            lca[0] = root
            
            # Recursively search
            if root.val < p.val and root.val < q.val:
                search(root.right)
            elif root.val > p.val and root.val > q.val:
                search(root.left)
            else:
                return
            
        # Call the function and return the res
        search(root)
        return lca[0]
            
        
```

## 2. Pure Recursion (纯函数式递归)
* **Key**: Divide and conquer without state mutation. Relies on the call stack to bubble the result up from the base case.
* **Thought Process**: Evaluate current node $\rightarrow$ if targets are strictly on one side, delegate the search to that subtree $\rightarrow$ if targets diverge, the current node is the LCA $\rightarrow$ return the result directly up the chain.
* **Pitfalls**: **Hidden Resource Drain**. Heavily relies on the OS call stack. Space complexity is $O(H)$, which degrades to $O(N)$ in an unbalanced tree, posing a Stack Overflow risk in highly constrained or concurrent environments.
* **Pattern**: Pure Recursive DFS (Mathematically elegant, but lacks resource restraint).


```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # Base Case
        if not root: return None
        
        # Iterate
        # Case1: p & q both at left subtree
        if root.val > p.val and root.val > q.val:
            return self.lowestCommonAncestor(root.left, p, q)

        # Case1: p & q both at right subtree
        if root.val < p.val and root.val < q.val:
            return self.lowestCommonAncestor(root.right, p, q)

        # Case1: p & q at both sides, current node is the lca
        else:
            return root
```

## 3. Iterative State Machine (迭代状态机 - 工业界最优解)
* **Key**: Exploits the "unidirectional determinism" of a BST. Eliminates the call stack entirely by using a single cursor to linearly traverse down the tree without backtracking.
* **Thought Process**: Allocate an $O(1)$ cursor in memory $\rightarrow$ initiate a `while` loop $\rightarrow$ adjust cursor left or right based on target values $\rightarrow$ immediately break the loop upon divergence and return the cursor.
* **Pitfalls**: Path dependence. Developers often default to recursion for tree problems, failing to recognize that a BST search is a one-way street requiring no historical state recovery.
* **Pattern**: Iterative State Machine / Pointer Traversal (The elite paradigm: absolute control over system resources with $O(1)$ space complexity).

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        while root:
            if p.val > root.val and q.val > root.val:
                root = root.right

            elif p.val < root.val and q.val < root.val:
                root = root.left

            else:
                return root

        return None
```
